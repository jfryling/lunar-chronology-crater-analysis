```python
import scipy
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pygmt

robbins = pd.read_csv("lunar_crater_database_robbins_2018.csv")
diams = pd.DataFrame(robbins, columns = ['DIAM_CIRC_IMG'])

fig, ax = plt.subplots()
ax.hist(diams, bins=100)
plt.yscale('log')
plt.xlabel('Diameter of Crater (km)')
plt.ylabel('Number of craters')
plt.show()
```

```python
diams_red = diams.loc[diams['DIAM_CIRC_IMG'] < 100]


fig, ax = plt.subplots()
ax.hist(diams_red, bins=100)
plt.yscale('log')
plt.xlabel('Diameter of Crater (km)')
plt.ylabel('Number of craters')
plt.show()
```
